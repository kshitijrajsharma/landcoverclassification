```python
import os 
```

```python
cwd = os.getcwd()
```

```python
data_path = os.path.join(cwd,'data','V2_final.geojson')
```

```python
import geopandas as gpd 
```

```python
data = gpd.read_file(data_path)
```

```python
data.to_crs(3857)
```

```python
data['centroid'] = data.centroid
```

```python
data.set_geometry('centroid')
```

```python
data['poly_geom'] = data['geometry']
```

```python
data['geometry'] = data['centroid']
```

```python
data.drop('centroid',axis=1)
```

```python
data_clean = data.drop('poly_geom',axis=1)
```

```python
data_clean = data.drop('centroid',axis=1)
```

```python
data_clean = data_clean.drop('poly_geom',axis=1)
```
